# Educacion en Mexico

```python
#Librerias usadas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

El conjunto de datos ha sido sacado de la pagina oficial del gobierno de la ciudad de Mexico. Dicho conjunto de datos fue posible gracias al trabajo de un poco más de 147 mil entrevistadoras y entrevistadores. <br>
Link: [Características educativas (Censo 2020)](https://datos.cdmx.gob.mx/dataset/caracteristicas-educativas-censo-2020)

### Datasets

```python
ed_mx = pd.read_csv('censo_c_educacion_loc_completo.csv')
```

```python
ed_mx.head()
```

```python
ed_mx1 = ed_mx.drop(['t_pob_3a5_noasiste', 't_pob_6a11_noasiste', 't_pob_12a14_noasiste',
                    't_pob_15a17_asiste', 't_pob_18a24_asiste', 't_pob_analfabeta',
                    't_pob_sin_escolaridad', 't_pob_prim_incomp', 't_pob_prim_comp',
                    't_pob_sec_incomp', 't_pob_sec_comp', 't_pob_educ_posbasica'], axis=1)
```

```python
ed_mx1
```

### Analizar el df y limpiarlo

```python
ed_mx1.isnull().sum()
```

No tenemos ningun dato faltante

```python
#Breve descripcion de que tipo de datos vamos a tratar
#y que columnas se encuentran en el dataframe
ed_mx1.info()
```

En cuanto a los datos numericos no tendremos ningun problema en tratarlos pues vienen ya en formato numerico

Como podemos observar tenemos varias instancias para la misma alcaldia lo que haremos sera crear un dataframe que contenga la suma de todos sus valores en una sola fila

```python
ed_mx_clean = ed_mx1.groupby('alcaldia')['pob_3a5_noasiste','pob_6a11_noasiste','pob_12a14_noasiste','pob_15a17_asiste',
                           'pob_18a24_asiste','pob_analfabeta','pob_sin_escolaridad','pob_prim_incomp',
                           'pob_prim_comp','pob_sec_incomp','pob_sec_comp','pob_educ_posbasica'].sum().reset_index()
```

```python
ed_mx_clean.head()
```

```python
ed_mx_clean.info()
```

Como podemos ver cambiamos de tener 91 instancias a solo 16

```python
#Agregar una nueva columna que contenga el total de niños que no asisten a la escuela
ed_mx_clean['t_ninos_noasisten'] = ed_mx_clean['pob_3a5_noasiste'] + ed_mx_clean['pob_6a11_noasiste'] + ed_mx_clean['pob_12a14_noasiste']
```

```python
ed_mx_clean.head()
```

### Visualizaciones

```python
#Poblacion de 3 a 5 que no va a la escuela por alcaldias
sns.barplot(data = ed_mx_clean, x = ed_mx_clean['alcaldia'], y = ed_mx_clean['pob_3a5_noasiste'])
plt.title('Cantidad de niños de 3-5 años que no asisten a la escuela \n')
plt.xlabel('Alcaldias')
plt.ylabel('No. de niños')
plt.xticks(rotation = 90)
```

```python
#Poblacion de 6 a 11 que no va a la escuela por alcaldias
sns.barplot(data = ed_mx_clean, x = ed_mx_clean['alcaldia'], y = ed_mx_clean['pob_6a11_noasiste'])
plt.title('Cantidad de niños de 6-11 años que no asisten a la escuela \n')
plt.xlabel('Alcaldias')
plt.ylabel('No. de niños')
plt.xticks(rotation = 90)
```

```python
#Poblacion de 12 a 14 que no va a la escuela por alcaldias
sns.barplot(data = ed_mx_clean, x = ed_mx_clean['alcaldia'], y = ed_mx_clean['pob_12a14_noasiste'])
plt.title('Cantidad de niños de 12-14 años que no asisten a la escuela \n')
plt.xlabel('Alcaldias')
plt.ylabel('No. de niños')
plt.xticks(rotation = 90)
```

```python
#Poblacion total de niños que no van a la escuela por alcaldias
sns.barplot(data = ed_mx_clean, x = ed_mx_clean['alcaldia'], y = ed_mx_clean['t_ninos_noasisten'])
plt.title('Cantidad de niños que no asisten a la escuela \n')
plt.xlabel('Alcaldias')
plt.ylabel('No. de niños que no asisten a la escuela')
plt.xticks(rotation = 90)
```

Observando el grafico anterior nos podemos percatar rapidamente de las alcaldias con mas cantidad de niños sin asistencia a la escuela.
* Iztapalapa
* Gustavo A. Madero
* Alvaro Obregon

```python
#Visualizar el top 10 alcaldias con mas numero de niños sin asistir a clases
top_alc_noasisten = ed_mx_clean.nlargest(10, ['t_ninos_noasisten']).set_index('alcaldia')['t_ninos_noasisten']
plt.figure(figsize=(10, 6))
sns.barplot(x = top_alc_noasisten, y = top_alc_noasisten.index)
plt.title('Las 10 alcaldias con mayor numero de niños sin asistir a la escuela \n')
plt.xlabel('Total de niños de 3-14 que no asisten a la escuela')
plt.ylabel('Alcaldias')
plt.savefig('pob_analfabeta_cmdx.png', bbox_inches = 'tight')
```
